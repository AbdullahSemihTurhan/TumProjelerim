```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.2,0.2,0.8,0.8])
eksen2=figure2.add_axes([0.2,0.2,0.4,0.4])#ilk 0.2 sola sabitliyor
#sonraki0.2 grafiğin sol tabanına sabitler
eksen1.plot(numpyDizisi1,numpyDizisi2,"r-*")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Baslik")
eksen2.plot(numpyDizisi2,numpyDizisi1,"m-*")
eksen2.set_xlabel("x ekseni kucuk")
eksen2.set_ylabel("y ekseni kucuk")
eksen2.set_title("kucuk grafik")
```

```python
(benimfigure,benimEksenler)=plt.subplots(nrows=1,ncols=2) #figure nin daha kolayi ancak figure deki gibi  detayli deil
#1 kolonlu 2 sirali cizelge olusturur 2 adet
#nrows = kolon adedi,ncols=sira adedi

for nedemekistersenonude in benimEksenler:
    nedemekistersenonude.plot(numpyDizisi1,numpyDizisi2,"g")
    nedemekistersenonude.set_xlabel("X ekseni")
plt.tight_layout() # 2 grafik arasina mesafe koyar dogru duzgun ayarlanmasina saglar
```

# gelişmiş gorsellestirme

```python
yeniFigur=plt.figure()

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2,label = "numpyDizisi**2") 
#label = ne verisi oldugunu daha rahat gormek icin bi detay altta bak

yeniEksen.legend()
```

ustteki ek veri ile yapilmis grafik,alttaki birden fazla veri ile yapilmis grafik ornegi

```python
yeniFigur=plt.figure()#dpi ayari yapabiliyoruz plt.figure(dpi=deger)
#figure(figsize=(uzunluk,genislik)) parantez icine  vererek girebiliyoruz ama girmesekde oluyor
 

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2,"g",label = "numpyDizisi**3") 
yeniEksen.plot(numpyDizisi1,numpyDizisi1**4,"r",label = "numpyDizisi**4") 
#numpy dizisi 1 in 4. kuvvetini aldirdim gormek icin
#label = ne verisi oldugunu daha rahat gormek icin bi detay altta bak
#renkler label kodundan once konur
yeniEksen.legend(loc=1) #label in yerini belirleyebilmek icin legend kullanilir

```

yeniFigur.savefig("denemefigurum.png",dpi=200) #seklinde png olarak kaydedebilirsin
